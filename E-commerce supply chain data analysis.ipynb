#1. Make a visualization of the price of the products and the revenue generated by them.

```python
fig = px.scatter(df, x = 'Price', y = 'Revenue generated', color = 'Product type', hover_data = ['Number of products sold'], trendline = 'ols')
fig.show()
```

#2. Analyze the Sales by Product Type.

```python
sales_data = df.groupby('Product type')['Number of products sold'].sum().reset_index()
sales_data
```

```python
pie_chart = px.pie(sales_data, values = 'Number of products sold',
                   names = 'Product type',
                   title = 'Sales by Product Type',
                   hover_data = ['Number of products sold'],
                   hole=0.5,
                   color_discrete_sequence=px.colors.qualitative.Pastel)
pie_chart.update_traces(textposition='inside', textinfo='percent+label')
pie_chart.show()
```

#3. Find out the total revenue generated from shipping carriers.

```python
df.head()
```

```python
total_revenue = df.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()
total_revenue
```

```python
fig = go.Figure()
fig.add_traces(go.Bar(x=total_revenue['Shipping carriers'],
                      y=total_revenue['Revenue generated']))
fig.update_layout(title='Total Revenue by Shipping Carrier',
                  xaxis_title='Shipping Carrier',
                  yaxis_title='Revenue Generated')
fig.show()
```

#4. Analyze the revenue generated by each SKU.

```python
df.head()
```

```python
revenue_chart = px.line(df, x='SKU',
                        y='Revenue generated',
                        title='Revenue Generated by SKU')
revenue_chart.show()
```

#5. Analyze the shipping cost of Carriers.

```python
df.head()
```

```python
shipping_cost_chart = px.bar(df, x='Shipping carriers',
                             y='Shipping costs',
                             title='Shipping Costs by Carrier')
shipping_cost_chart.show()
```

#6. Find out the cost distribution by transportation mode.

```python
df.head()
```

```python
transportation_chart = px.pie(df, values='Costs',
                              names='Transportation modes',
                              title='Cost Distribution by Transportation Mode',
                              hole=0.5,
                              color_discrete_sequence=px.colors.qualitative.Pastel)
transportation_chart.show()
```

#7. Analyzing the Defect Rate of the product during shipping.

```python
df.head()
```

```python
defect_rates_by_product = df.groupby('Product type')['Defect rates'].mean().reset_index()
defect_rates_by_product
```

```python
fig = px.bar(defect_rates_by_product, x='Product type', y='Defect rates', title='Average Defect Rates by Product Type')
fig.show()
```
